```python
# Exemplo para colorir nós de cores diferentes
import networkx as nx

G=nx.Graph()

G.add_node('tatuape')
G.add_node('butanta')
G.add_node('bras')
# G.add_node('piracicaba')

G.add_edge('tatuape', 'butanta')
G.add_edge('tatuape', 'bras')
# G.add_edge('bras','piracicaba')

H=nx.Graph()
H.add_node('petrobras')
H.add_node('unisinos')
H.add_node('canoas')
# H.add_node('piracicaba')

H.add_edge('petrobras', 'unisinos')
H.add_edge('unisinos', 'canoas')
# H.add_edge('canoas', 'piracicaba')

F = nx.compose(G,H)
F.add_node('piracicaba')

F.add_edge('canoas', 'piracicaba')
F.add_edge('bras', 'piracicaba')

print(G.nodes())
print(H.nodes())
print(F.nodes())

print(G.edges())
print(H.edges())
print(F.edges())

# pos = nx.spring_layout(G, seed=28)
# nx.draw_networkx(G, pos, node_color="yellow")

# pos = nx.spring_layout(H, seed=38)
# nx.draw_networkx(H, pos, node_color="green")

color_map = []
for node in F.nodes():
    if node in ['tatuape', 'butanta', 'bras']: # criar um array com o nome das estacoes da linha vermelha, verde, etc
        color_map.append('blue')
    elif node in ['piracicaba']: # baldeações
        color_map.append('grey')
    else: 
        color_map.append('green')      
# nx.draw(F, node_color=color_map)
# plt.show()

pos = nx.spring_layout(F, seed=48)
nx.draw_networkx(F, pos, node_color=color_map)
```
